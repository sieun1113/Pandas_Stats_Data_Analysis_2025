###**<font color = 'red'> ğŸ¹ğŸ¾ Read and write the csv format**

```python
# csv (comma separated values) format

import pandas as pd

#file directory
file = "/content/sample_data/sample.csv"

#read_csv() functionìœ¼ã„¹ ì‚¬ìš©í•˜ì—¬ ë°ì´í„°í”„ë ˆì„ ë³€í™˜, ê¸°ë³¸ê°’ 0í–‰ì„ ì—´ ì§€ì • (ì—´ì§€ì •ì´ ë­ì•¼??)
df1 = pd.read_csv(file)
print(df1)
print('\n')

#header = None ì„ íƒ
df2 = pd.read_csv(file, header = None)
print(df2)
print('\n')

#hearder = 1 ì„ íƒ
df3 = pd.read_csv(file, header = 1)
print(df3)
print('\n')

#index_col = False (ì¸ë±ìŠ¤ ì§€ì •í•˜ì§€ ì•ŠìŒ)
df4 = pd.read_csv(file, index_col = False)
print(df4)
print('\n')

#index_col = 'c0' ì²« ë²ˆì§¸ ì—´ì„ ì¸ë±ìŠ¤ ì§€ì •
df5 = pd.read_csv(file, index_col = 'c0')
print(df5)


```

```python
import pandas as pd

url = "/content/sample_data/sample.html"

#html ì›¹í˜ì´ì§€ì˜ í‘œë¥¼ ê°€ì ¸ì™€ì„œ ë°ì´í„°í”„ë ˆì„ìœ¼ë¡œ ë³€í™˜
tables = pd.read_html(url)

#í‘œ ê°œìˆ˜ í™•ì¸
print(len(tables))
print('\n')

#í‘œ ë¦¬ìŠ¤íŠ¸ì˜ ì›ì†Œë¥¼ ë°˜ë³µí•˜ë©´ì„œ (interating the element of the table list) í™”ë©´ ì¶œë ¥
for i in range(len(tables)):
  print("tables[%s]" %i)
  print(tables[i])
  print('\n')

#íŒŒì´ì¬ íŒ¨í‚¤ì§€ ì •ë³´ê°€ ë“¤ì–´ ìˆëŠ” ë‘ ë²ˆì§¸ ë°ì´í„°í”„ë ˆì„ì„ ì„ íƒí•˜ì—¬ df ë³€ìˆ˜ì— ì €ì¥
df = tables[i]

#name ì—´ì„ ì¸ë±ìŠ¤ë¡œ ì €ì¥
df.set_index(['name'], inplace = True)
print(df)
```
