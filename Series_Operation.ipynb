### ğŸ¹ Pandas Series vs. Series
- ì‹œë¦¬ì¦ˆì˜¤ ì‹œë¦¬ì¦ˆ ì‚¬ì´ì— ì‚¬ì¹™ì—°ì‚° (e.g., addition, subtraction, multiplication, division)ì„ ì²˜ë¦¬í•˜ëŠ” ë°©ë²•

```python
import pandas as pd

student1 = pd.Series({'English':88, 'Literature':90, 'Math':98})
student2 = pd.Series({'Literature':78, 'Math':90, 'English':'80'})
print(student1)
print('\n')
print(student2)
print('\n')

#ë‘ í•™ìƒì˜ ì ìˆ˜ë¡œ ì‚¬ì¹™ì—°ì‚° ìˆ˜í–‰ í•´ ë³´ê¸°

student1 = student1.astype(int)
student2 = student2.astype(int)

addition = student1 + student2
print(addition)
print('\n')

subtraction = student1-student2
print(subtraction)
print('\n')

multiplication = student1*student2
print(multiplication)
print('\n')

division = student1/student2
print(division)
print('\n')

result = pd.DataFrame([addition, subtraction, multiplication, division], index = ['addition +', 'subtraction -', 'multiplication *', 'division /'])
print(result)
```
