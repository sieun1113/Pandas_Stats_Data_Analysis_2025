####<font color = 'red'> ğŸ‘€ğŸ¾ **ë°ì´í„°í”„ë ˆì„ ë³€í™˜ Converting to basic data frame from dictionary data** 1 â¤µï¸

```python
#-*- coding: utf-8 -*-

import pandas as pd

#DataFrame( ) í•¨ìˆ˜ë¡œ ë°ì´í„°í”„ë ˆì„ ë³€í™˜, ë³€ìˆ˜ dfì— ì €ì¥

dict_data = {'c0' : [1,2,3],
             'c1' : [4,5,6],
             'c2' : [7,8,9],
             'c3' : [10,11,12],
             'c4' : [13,14,15]}

df = pd.DataFrame(dict_data)
print(df)
print('\n')
```

####<font color = 'red'> ğŸ‘€ğŸ¾ **í–‰ ì¸ë±ìŠ¤ ì‹¤í–‰í•˜ê¸° set_index( ) 2-1** â¤µï¸

```python
#Dictionaryë°ì´í„°ë¥¼ í”„ë ˆì„ìœ¼ë¡œ ë³€í™˜í•˜ë©°, í–‰ ì¸ë±ìŠ¤ë¥¼ 'r0', 'r1', 'r2'ë¡œ ì¬ì§€ì •

df = pd.DataFrame(dict_data, index=['r0', 'r1', 'r2'])
print(df)
print('\n')
```

####<font color = 'red'> ğŸ‘€ğŸ¾ **#í–‰(row) ì¸ë±ìŠ¤ ì¬ë°°ì—´ reindex( ) 2-2** â¤µï¸

```python
#ì¸ë±ìŠ¤ë¥¼ [r0, r1, r2, r3, r4]ë¡œ ì¬ì§€ì •
new_index = ['r0','r1','r2','r3','r4']
ndf = df.reindex(new_index)
print(ndf)
print('\n')
```

####<font color = 'red'> ğŸ‘€ğŸ¾ **#reindex ( )ë¡œ ë°œìƒí•œ NaN ê°’ì„ ìˆ«ì 0ìœ¼ë¡œ ì±„ìš°ê¸° 3** â¤µï¸

```python
new_index = ['r0','r1','r2','r3','r4']
ndf2 = df.reindex(new_index, fill_value=0)
print(ndf2)
print('\n')
```

####<font color = 'red'> ğŸ‘€ğŸ¾ **í–‰(row) ì¸ë±ìŠ¤ ì´ˆê¸°í™” reset_index( ) 4** â¤µï¸

```python
ndf2=df.reset_index()
print(ndf2)
```

####<font color = 'red'> ğŸ‘€ğŸ¾ **í–‰(row) ì¸ë±ìŠ¤ë¥¼ ê¸°ì¤€ìœ¼ë¡œ ë°ì´í„° ì •ë ¬ sort_index( ) 5** â¤µï¸

```python
ndf3 = df.sort_index()
print(ndf3)
```
